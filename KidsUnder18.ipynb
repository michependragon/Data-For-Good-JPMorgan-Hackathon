```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # Import numpy for NaN handling

csv_file_path = 'Nonprofit_Data_PA.csv'

df = pd.read_csv(csv_file_path)

# Replace 'N' with NaN in specific columns
cols_to_replace = ['n10_19', 'n5_9', 'n<5']
df[cols_to_replace] = df[cols_to_replace].replace('N', np.nan)

# Convert columns to numeric
df['n10_19'] = pd.to_numeric(df['n10_19'], errors='coerce')  # Convert 'n10_19' column to numeric
df['n5_9'] = pd.to_numeric(df['n5_9'], errors='coerce')  # Convert 'n5_9' column to numeric
df['n<5'] = pd.to_numeric(df['n<5'], errors='coerce')  # Convert 'n<5' column to numeric

filtered_df = df[(df['zip'] >= 19101) & (df['zip'] <= 19154)]

total = filtered_df[['n10_19', 'n5_9', 'n<5']].sum(axis=1, skipna=True)

#print(total)

# Plotting the bar chart here
plt.figure(figsize=(10, 5))
plt.bar(filtered_df['zip'], total, color='skyblue')


plt.xlabel('ZipCode')
plt.ylabel('Total Kids Under 19')
plt.title('Number of Kids per Zip Code')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
